```python
# Review the distribution of your data with describer() function
df.describe()
```

* **Count:** refers to the number of non-missing (non-null) values in each column.
* **Mean:** represents the average. Provides a measure of central tendency, indicating the "typical" value in the dataset.
* **Standard deviation:** provides a measure of how spread out the values in a dataset are. If the standard deviation is low, it indicates that the values tend to be close to the mean. If the standard deviation is high, it indicates that the values are more spread out from the mean.
* **Minimum:** is the lowest value in the dataset.
* **Maximum:** is the highest value in the dataset.
* **Percentiles:** help in understanding the spread and distribution of the data, especially in large datasets where summarizing them with just the mean and standard deviation may not provide enough insight into the data's variability.

```python
# Median
print(df.median())
```

```python
# Plot the distirbution of ph
column_to_visualize = 'ph'

# Plotting a histogram
plt.figure(figsize=(6, 4))
plt.hist(df[column_to_visualize], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of {}'.format(column_to_visualize))
plt.xlabel(column_to_visualize)
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot the distribution for all columns
plt.figure(figsize=(6, 4))
for column in df.columns:
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title('Histogram of {}'.format(column))
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()  # Display each histogram before moving to the next column
```

```python
# Calculate pair-wise correlation between variables using the corr() function
df.corr(numeric_only = True) # Ignore any non-numeric values
```

* Computes pairwise correlation of columns, excluding null values.
* The correlation coefficient ranges from -1 to 1:
    * The correlation of a variable with itself is 1.
    * 0 indicates no linear relationship.
    * Close to +1 indicates a strong positive correlation. It suggests that as one variable increases, the other variable tends to increase as well.
    * Close to -1 indicates a strong negative correlation. It implies that the two variables move in opposite directions, meaning when one variable increases, the other tends to decrease, and vice versa.
* Methods:
    * pearson: standard correlation coefficient (Default method)
    * 
kendall: Kendall Tau correlation coefficie
    *  
spearman: Spearman rank correlationion

```python
df.corr(method='pearson')
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```
