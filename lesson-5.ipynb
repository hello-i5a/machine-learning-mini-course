# Day 5: Understand Data with Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/master/winequality-white.csv"
df = pd.read_csv(url)
print(df)
```

```python
df = pd.read_csv(url, delimiter=';') # By default, this parameter is set to `,` (comma)
print(df)
```

```python
# Use the hist() function to create a histogram of each attribute
df.hist(bins=25, grid=False, figsize=(10,8), color='#86bf91')

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

**Histograms**
* Help illustrate the distribution of the data.
* Help detect any outliers or gaps in the data.

```python
# Use the plot(kind='box') function to create box and whisker plots of each attribute
df.plot.box(figsize = (10, 8), title = "Whisker Plots for Each Attribute")

# Adjust x-axis label font size and rotation
plt.xticks(fontsize=12, rotation=45)
plt.show()
```

**Box and Whisker Plots**
* Shows the outliers, minimum and maximum data values, lower and upper quartile values, as well as the median at a glance.

```python
# Use the pandas.scatter_matrix() function to create pairwise scatter plots of all attributes
scatter_matrix = pd.plotting.scatter_matrix(df, figsize=(10, 8))

# Adjust text and axis labels
for ax in scatter_matrix.flatten():
    ax.xaxis.label.set_rotation(45)
    ax.yaxis.label.set_rotation(45)
    ax.yaxis.label.set_ha('right') # Set the horizontal alignment of the y-axis labels to the right for better readability

plt.tight_layout()
plt.show()
```

**Pairwise Scatter Plots**
* Shows the relationship between any two variables of the concerned dataset.
